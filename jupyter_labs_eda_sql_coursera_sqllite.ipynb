### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
sql_statement = '''
    SELECT DISTINCT LAUNCH_SITE
    FROM SPACEXTBL
'''
# Execute the SQL statement and fetch the result into a DataFrame
Problem1 = pd.read_sql(sql_statement, conn)
Problem1
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA'


```python
sql_statement = '''
    SELECT *
    FROM SPACEXTBL
    WHERE LAUNCH_SITE LIKE 'CCA%'
    LIMIT 5
'''
# Execute the SQL statement and fetch the result into a DataFrame
Problem2 = pd.read_sql(sql_statement, conn)
Problem2
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
sql_statement = '''
    SELECT sum(PAYLOAD_MASS__KG_) AS total_payload_mass
    FROM SPACEXTBL
    WHERE Customer ='NASA (CRS)'
'''
# Execute the SQL statement and fetch the result into a DataFrame

Problem3 = pd.read_sql(sql_statement, conn)
Problem3
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
sql_statement = '''
    SELECT avg(PAYLOAD_MASS__KG_) AS Average_payload_mass
    FROM SPACEXTBL
    WHERE Booster_Version ='F9 v1.1'
'''
# Execute the SQL statement and fetch the result into a DataFrame

Problem4 = pd.read_sql(sql_statement, conn)
Problem4
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_


```python
sql_statement = '''
    SELECT min(Date)
    FROM SPACEXTBL
    WHERE Landing_Outcome ='Success (ground pad)'
    ORDER BY Date
'''
# Execute the SQL statement and fetch the result into a DataFrame

Problem5 = pd.read_sql(sql_statement, conn)
Problem5
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
sql_statement = '''
    SELECT Booster_Version
    FROM SPACEXTBL
    WHERE PAYLOAD_MASS__KG_>4000
    AND PAYLOAD_MASS__KG_<6000
    AND Landing_Outcome='Success (drone ship)'
'''
# Execute the SQL statement and fetch the result into a DataFrame

Problem6 = pd.read_sql(sql_statement, conn)
Problem6
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
sql_statement = '''
    SELECT count(Mission_Outcome) AS total_Mission_Outcome
    FROM SPACEXTBL
    WHERE MISSION_OUTCOME = 'Success' or MISSION_OUTCOME = 'Failure'
'''
# Execute the SQL statement and fetch the result into a DataFrame

Problem7 = pd.read_sql(sql_statement, conn)
Problem7
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
sql_statement = '''
    SELECT Booster_Version
    FROM SPACEXTBL
    WHERE PAYLOAD_MASS__KG_ =(SELECT max(PAYLOAD_MASS__KG_)
                            FROM SPACEXTBL)
'''
# Execute the SQL statement and fetch the result into a DataFrame

Problem8 = pd.read_sql(sql_statement, conn)
Problem8
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 4, 2) as month to get the months and substr(Date,7,4)='2015' for year.**


```python
sql_statement = '''
    SELECT substr(Date, 4, 2) AS Month,Booster_Version,Launch_Site
    FROM SPACEXTBL
    WHERE Landing_Outcome='Failure (drone ship)'
    AND substr(Date,7,4)='2015'
'''
# Execute the SQL statement and fetch the result into a DataFrame

Problem9 = pd.read_sql(sql_statement, conn)
Problem9
```

### Task 10




##### Rank the  count of  successful landing_outcomes between the date 04-06-2010 and 20-03-2017 in descending order.


```python
sql_statement = '''
    SELECT Date,Landing_Outcome
    FROM SPACEXTBL
    WHERE Landing_Outcome='Failure (drone ship)'
    AND substr(Date,7,4)='2015'
'''
# Execute the SQL statement and fetch the result into a DataFrame

Problem9 = pd.read_sql(sql_statement, conn)
Problem9
```
